```python
#по отдельности выводим первый эл массива, последний и пятый
print(test_ar[0]) 
print(test_ar[-1])
print(test_ar[4])
```

```python
print(test_ar[2:6]) #со второго по пятый эл
print(test_ar[::2]) # каждый второй эл
print(test_ar[3:]) # с третьего эл и до конца
print(test_ar[:2]) #с начала и до первого эл 
```

```python
a = np.array([2, 5, 3, 6, 7, 4, 0, 1, 4, 5, 6, 8])
a
```

```python
print(np.mean(a)) #среднее
print(np.median(a)) #медиана
print(np.std(a)) #стандартное отклонение
print(np.var(a)) #дисперсия
print(np.min(a)) #минимальное
print(np.max(a)) #максимальное
print(np.sum(a)) #сумма
```
